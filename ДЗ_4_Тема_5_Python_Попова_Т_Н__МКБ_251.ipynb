```python
from datetime import datetime
magazines = ["The Moscow Times - Wednesday, October 2, 2002", "The Guardian - Friday, 11.10.13", "Daily News - Thursday, 18 August 1977"]
sub = "- "
res = []
for i in range(len(magazines)):
    res.append(magazines[i].split(sub)[1])

dt1 = datetime.strptime(res[0], "%A, %B %d, %Y")
print(dt1)
# print(type(dt1))

dt2 = datetime.strptime(res[1], "%A, %d.%m.%y")
print(dt2)
# print(type(dt2))

dt3 = datetime.strptime(res[2], "%A, %d %B %Y")
print(dt3)
# print(type(dt3))
```

```python
from datetime import datetime, timedelta

def date_range(start, end):
    result = []

    try:
        d1 = datetime.strptime(start, "%Y-%m-%d")
        d2 = datetime.strptime(end, "%Y-%m-%d")
    except:
        return []

    if d1 > d2:
        return []

    while d1 <= d2:
        result.append(d1.strftime("%Y-%m-%d"))
        d1 += timedelta(days=1)

    return result

# print(date_range("2022-01-01", "2022-01-03"))
# print(date_range("2022-01-03", "2022-01-01"))
# print(date_range("2022-02-30", "2022-02-31"))
```
